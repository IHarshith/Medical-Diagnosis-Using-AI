```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import seaborn as sns
```

```python
df = pd.read_csv('/content/kidney_disease.csv')
df
```

```python
df['pcv'].unique()
```

```python
df.replace(
    {
    'ckd':1,
    'notckd':0,
    'yes':1,
    'no':0,
    'normal':0,
    'abnormal':1,
    'present':1,
    'notpresent':0,
    'ckd\t':1,
    'poor':1,
    'good':0,
    '\tyes':1,
    '\tno':0,
    '\t?':np.nan,
    '\t43':43
    },
    inplace=True
)
```

```python
df.dtypes
```

```python
#To transform object types into integer or float and ignoe=re errors

cols_to_replace = ['pcv','wc','rc','dm']
df[cols_to_replace] = df[cols_to_replace].apply(pd.to_numeric, errors='coerce')
```

```python
df.dtypes
```

```python
df.fillna(df.mean(), inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum().sum()
```

#Training and Testing

```python
x=df.drop(['classification'],axis=1)
y=df['classification']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=45)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()

lr.fit(x_train,y_train)

y_pred = lr.predict(x_test)

accuracy = accuracy_score(y_test,y_pred)
print(accuracy)
```

```python
x_train = x_train.drop(['id','sg','rbc','pc','pcc','ba','sc','sod','pot','hemo','rc','appet','pe'],axis=1)
x_test = x_test.drop(['id','sg','rbc','pc','pcc','ba','sc','sod','pot','hemo','rc','appet','pe'],axis=1)
```

```python
x_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
x_train_prediction = lr.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
print("The training accuracy: ",training_data_accuracy)
```

```python
x_test_prediction = lr.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
print("The test accuracy: ",test_data_accuracy)
```
