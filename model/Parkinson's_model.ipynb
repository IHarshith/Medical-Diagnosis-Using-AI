```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

```python
df = pd.read_csv('/content/parkinson_data.csv')
df
```

```python
from sklearn.svm import SVC
```

```python
x = df.drop(['name','status'],axis=1)
y =df['status']
```

```python
x.head()
```

```python
y.head()
```

```python
x.shape,y.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.25,random_state=0)
```

```python
svm = SVC()
svm.fit(x_train,y_train)
```

```python
test_pred = svm.predict(x_test)
test_pred_acc = accuracy_score(y_test,test_pred)
print('The testing accuracy is: ',test_pred_acc)
```

```python
train_pred = svm.predict(x_train)
train_pred_acc = accuracy_score(y_train,train_pred)
print('The Training Accuracy is : ',train_pred_acc)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
l_train_pred = lr.predict(x_train)
l_train_accuracy = accuracy_score(y_train,l_train_pred)
print('The Training accuracy of Logistic Regression is :',l_train_accuracy)
```

```python
l_test_pred = lr.predict(x_test)
l_test_acc = accuracy_score(y_test,l_test_pred)
print('The testing accuracy is :',l_test_acc)
```
